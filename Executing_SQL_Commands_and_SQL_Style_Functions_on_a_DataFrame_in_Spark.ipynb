### Executing SQL Commands and SQL-Style Functions on a DataFrame in Spark

```python
from pyspark.sql import SparkSession
from pyspark.sql import Row
import collections
```

```python
# Create a SparkSession (Note, the config section is only for Windows!)
spark = SparkSession.builder.config("spark.sql.warehouse.dir",
"file:///C:/temp").appName("SparkSQL").getOrCreate()
```

```python
def mapper(line):
  fields = line.split(',')
  return Row(ID=int(fields[0]),
name=str(fields[1].encode("utf-8")), age=int(fields[2]),
numFriends=int(fields[3]))
lines = spark.sparkContext.textFile("fakefriends.csv")
people = lines.map(mapper)
```

```python
schemaPeople = spark.createDataFrame(people).cache()
schemaPeople.createOrReplaceTempView("people")
```

```python
teenagers = spark.sql("SELECT * FROM people WHERE age >= 13 AND age <= 19")
```

```python
for teen in teenagers.collect():
  print(teen)
```

```python
schemaPeople.groupBy("age").count().orderBy("age").show()
```

```python
spark.stop()
```
